Testing the model

```python
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression

import statsmodels.api as sm

from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
prices=df2['price']
features=df2.drop(['price','date'],axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, prices, test_size = 0.2, random_state = 10)
```

```python
X_train.head()
```

```python
features.info()
```

```python
regr = LinearRegression()
regr.fit(X_train, y_train)
```

```python
regr.coef_
```

```python
regr.intercept_
```

```python
pd.DataFrame(data =regr.coef_, index = X_train.columns, columns = ['COEF'] )
```

```python
print('The score of the Regresson with the training data is ', regr.score(X_train, y_train))
```

```python
print('The score of the Regresson on  the testing data is ', regr.score(X_test, y_test))
```
